```python
import nbformat
from nbconvert import MarkdownExporter

# Define input and output paths
input_path = '/content/drive/MyDrive/Python_for_ESADE/Summary/Cheat_sheet_pandas_&_plotting.ipynb'
output_path = '/content/drive/MyDrive/Python_for_ESADE/Summary/Cheat_sheet_pandas_&_plotting.md'

try:
    # Load the notebook
    with open(input_path, "r", encoding="utf-8") as f:
        nb = nbformat.read(f, as_version=4)

    # Convert to markdown
    exporter = MarkdownExporter()
    body, resources = exporter.from_notebook_node(nb)

    # Write to .md file
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(body)

    print("‚úÖ Notebook successfully converted to Markdown!")
except Exception as e:
    print("‚ùå Error during conversion:", str(e))
```
