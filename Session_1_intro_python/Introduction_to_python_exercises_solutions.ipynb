#### ðŸ‹ Exercise 1 : Lists

Define the following list:

`number_list = [1,4,25,32,-1,0,12,32]`

  - a) Give the element in position 2

  - b) Give the last element of the list

  - c) Which is the length of the list

  - d) How many elements are greater than 10 (A bit more advanced)

  - e) Add all the elements of the list. Which is the result?

  - f) Append the number 14 to the list

````python
#### Exercise 1

# ðŸ“‹ List Practice with f-strings: Indexing, Length, Filtering, Summing, and Appending

# Given list
number_list = [1, 4, 25, 32, -1, 0, 12, 32]

# a) Get the element at position 2
element_pos_2 = number_list[2]
print(f"a) Element at position 2: {element_pos_2}")

# b) Get the last element of the list
last_element = number_list[-1]
print(f"b) Last element of the list: {last_element}")

# c) Get the length of the list
length_of_list = len(number_list)
print(f"c) Length of the list: {length_of_list}")

# d) Count how many elements are greater than 10
count_greater_than_10 = 0
for num in number_list:
    if num > 10:
        count_greater_than_10 += 1
print(f"d) Elements greater than 10: {count_greater_than_10}")

# e) Sum all the elements in the list
total_sum = sum(number_list)
print(f"e) Sum of all elements: {total_sum}")

# f) Append the number 14 to the list
number_list.append(14)
print(f"f) Updated list after appending 14: {number_list}")
````

You're given a list of numbers (E.g `[3,8,12,5,7]`). Your task is to:

1. Loop through each number in the list.
2. For each number:
   - Print the number.
   - Print whether it's **even or odd**.

Use a `for` loop and the `%` operator to help determine evenness.

E.g if input list is `[1,3,10]` expected output is:

```
1 is odd

```



````python
#Solution
numbers = [3, 8, 12, 5, 7]

for number in numbers:
    if number % 2 == 0:
        print(f"{number} is even.")
    else:
        print(f"{number} is odd.")
````

#### ðŸ¤” Exercise 2

Write a Python function that takes a list and returns a new list with the distinct elements from the first list.

- Named the function `get_unique_elements`. The function it will take as input the argument `input_list`. The output will be a list with the unique elements.

For example if

- Input List : [1,2,3,3,3,3,4,5]

get_unique_elements will return:

- Output List : [1, 2, 3, 4, 5]

````python
#Solution
def get_unique_elements(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

# ðŸ” Test it
sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
result = get_unique_elements(sample_list)
print("Unique List:", result)
````

#### ðŸ¦ Exercise 1

Youâ€™re given a list that contains a mix of strings and numbers:

`list_mix = ['Horse', 'Capibara', 10, -3, 'Steak', 31, 'Eagle', 'Tortoise', -1]`

Your task is to:

- Create a new list called `list_strings` that contains only the string elements from `list_mix`.

- Create another list called `list_nums` that contains only the numeric elements (integers or floats) from `list_mix`.

Use a loop and `isinstance()` (or `type()`) to check the type of each element. Try to solve it without using any external libraries.

````python
# ðŸŽ¯ Mixed list to split into strings and numbers
list_mix = ['Horse', 'Capibara', 10, -3, 'Steak', 31, 'Eagle', 'Tortoise', -1]

# Create empty lists for strings and numbers
list_strings = []
list_nums = []

# Loop through the mixed list and classify elements
for item in list_mix:
    if isinstance(item, str):
        list_strings.append(item)
    elif isinstance(item, (int, float)):
        list_nums.append(item)

# Print results using f-strings
print(f"Original mixed list: {list_mix}")
print(f"List of strings: {list_strings}")
print(f"List of numbers: {list_nums}")
````
