### Section 4: Control Flow

In Python, control flow refers to the order in which individual instructions, statements, or function calls are executed or evaluated. It determines how the program makes decisions, repeats actions, and executes code conditionally or sequentially.



##### 4.1.Conditional statement

In Python, conditional statements allow your program to make similar decisions and take different paths depending on whether a condition is true or false.

üß† Keywords you need to know:

- `if`: test a condition ‚Äî if it's True, run a block of code.

- `elif` (short for "else if"): test another condition if the first one wasn't true.

- `else`: catch-all when no previous conditions matched.

``````python
# Conditionals
number = 5
if number > 0:
    print("Positive")
elif number == 0:
    print("Zero")
else:
    print("Negative")
``````

In this example, notice how the code inside each if, elif, and else block is indented (moved slightly to the right).
**Indentation** is not just for style in Python ‚Äî it is mandatory.
It tells Python which lines of code belong to each block.
Without proper indentation, Python will not understand your code and will raise an error

Sometimes you need to check more than one condition at the same time.
In Python, you can combine conditions using logical operators:

üß† Logical operators you need to know:

- `and`: both conditions must be True.

- `or`: at least one condition must be True.

- `not`: reverses the truth value of a condition.

``````python
age = 25
country = "Spain"
if age >= 18 and country == "Spain":
    print("Adult living in Spain")
elif age < 18 and country == "Spain":
    print("Minor living in Spain")
elif  country !="Spain":
    print("Not living in Spain")
else:
    print("Unknown")
``````

In this example:

- First, the program checks if both age >= 18 and country == "Spain" are true.

- If yes, it prints "Adult living in Spain".

- If not, it checks if both age < 18 and country == "Spain" are true.

- If yes, it prints "Minor living in Spain".

- If neither of the above are true, it checks if country != "Spain" (meaning the person does not live in Spain).

- If yes, it prints "Not living in Spain".

- If none of the previous conditions are met (very unlikely in this case), the program will go to the final else block and print "Unknown".



#### 4.2.Loops

Loops are used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string)

##### 4.2.1 For loops

A for loop lets you repeat an action for each item in a collection, like a list. It's one of the most powerful tools for automation and data processing.

Basic structure is

````
for item in collection:
    # Do something with item
````

- `item` is a temporary variable that represents the current element in the loop.

- `collection` can be a list, string, or anything you can iterate over.

Let's see an example

``````python
# Loops
list_numbers = [1, 2, 3, 4, 5]
for num in list_numbers:
    print(num)
``````

To loop through a set of code a specified number of times, we can use the `range()` function.

The `range()` function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

``````python
for i in range(5):
    print(i)
``````

##### 4.2.2 While loops

A while loop lets you repeat a block of code as long as a condition remains true. Unlike for loops (which are used when you know how many times to loop), while loops are useful when you're not sure how many times something should run ‚Äî only that it should run until a condition changes.

Basic Structure:

````
while condition:
    # Do something
`````

What's happening:
  
  - The loop checks the condition.

  - If the condition is `True`, it runs the block of code.

  - After each loop, it checks the condition again.

  - If the condition becomes `False`, it exits the loop.



``````python
# While loop (executes as long as a condition is true)
counter = 0
while counter < 3:
    print("Meow")
    counter = counter + 1
``````

###### üìà **Matplotlib**

Matplotlib is Python's most popular library for creating static, animated, and interactive plots.  It lets you visualize data using line charts, bar charts, scatter plots, and more. Highly customizable and works well with NumPy and Pandas.  
üîó [Matplotlib Docs](https://matplotlib.org/stable/contents.html)

``````python
import matplotlib.pyplot as plt

# Plot a line graph
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

plt.plot(x, y)
plt.title("My First Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
``````

``````python
# Let's do now a scatter plot

x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]
plt.scatter(x, y, label='Scatter Plot', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Scatter Plot')
plt.legend()
plt.show()
``````

``````python
# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]
plt.bar(categories, values, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Plot')
plt.show()
``````
