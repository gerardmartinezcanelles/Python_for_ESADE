2.6) Create a new column such as if date is 2025-04-05 or 2025-04-06 then it takes value'Weekend' else 'Workday'. Named the colum `day_type`. Steps:

- Create a function named `work_weekend()` that takes as input a date and performs:

  - if date is equal '2025-04-05' o '2025-04-06' it should return 'Weekend'

  - else it should return 'Workday'

- Use apply() method over df['date'] column.

`````python
def work_weekend(x):
  if x == '2025-04-05' or x == '2025-04-06':
    return 'Weekend'
  else:
    return 'Workday'
`````

`````python
df['day_type'] = df['date'].apply(work_weekend)
`````

2.7) Create a pivot table such that it computes the total number of views per `day_type`.

`````python
df_days = pd.pivot_table(df,index='day_type',values='views',aggfunc='sum')
df_days
`````

2.9) For each country is there any preference for a specific platform? To do so, create a pivot table and sum the number of unique users per country and platform.  Store the results in a dataframe called df_ct_plat

`````python
df_ct_plat = pd.pivot_table(df,index='country',columns='platform',values='unique_users',aggfunc='sum')
df_ct_plat
`````

2.10) (BONUS) In the previous DataFrame `df_ct_plat`, the values represent the absolute number of unique users per platform and country.

Now, compute the relative frequency (as percentages) of platform usage within each country.

In other words, for each country, calculate what percentage of its total unique users chose each platform.

Store the resulting DataFrame in a new variable called `df_ct_plat_pct`


ðŸ’¡ Hint:
You can use .div() along with .sum(axis=1) to divide each row by its row total.
For example:

  `df_ct_plat.div(df_ct_plat.sum(axis=1),axis=0)*100`


Note that at the end, we are computing the market share per country for each platform

`````python
df_ct_plat_pct  = df_ct_plat.div(df_ct_plat.sum(axis=1),axis=0)*100
df_ct_plat_pct
`````
