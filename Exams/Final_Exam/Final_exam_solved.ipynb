### Aggregations

2.1) Create a pivot table to count how many `Unique_users` are in each `Region`. Store the results in a dataframe called `df_region`

ðŸ’¡ *Hint: Recall the syntax for `pd.pivot_table(df,index=,values=,aggfunc=)`

````python
df_region = pd.pivot_table(df,index='Region',values='Unique_users',aggfunc='sum')
df_region
````

````python
df_region['relative_frequency'] = (df_region['Unique_users']/df_region['Unique_users'].sum())*100
df_region
````
