if amount < 50 then low

if amount>=50 and amount<100 then 'med'

if amount>=100 'High'

1.- def funtion expense_classifier

2.- apply method

```python
def expense_classifier(number):
  if number<50:
    return 'Low'
  elif number>=50 and number<100:
    return 'Mid'
  else:
    return 'High'
```

```python
df_exp['amount'].head()
```

```python
df_exp['amount_in_buckets'] = df_exp['amount'].apply(expense_classifier)
df_exp
```
