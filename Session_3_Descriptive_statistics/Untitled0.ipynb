```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_dataset= '/content/drive/MyDrive/Python_for_ESADE/Session_3_Descriptive_statistics/datasets/player_base.xlsx'
df_plot = pd.read_excel(path_dataset)
```

```python
df_plot
```

```python
plt.figure()
plt.title('Boxplot moneyspent')
plt.boxplot(df_plot['moneyspent'])
plt.xlabel('Moneyspent')
plt.show()
```

```python
plt.figure()
plt.title('Boxplot moneyspent')
plt.hist(df_plot['moneyspent'])
plt.xlabel('Moneyspent')
plt.show()
```

```python
plt.figure()
plt.title('Boxplot moneyspent')
sns.histplot(df_plot['moneyspent'])
plt.xlabel('Moneyspent')
plt.show()
```

```python
df_plot.head()
```

```python
plt.figure()
plt.title('Scatter plot')
plt.scatter(df_plot['gameroundsplayed'], df_plot['levelscompleted'])
plt.xlabel('gameroundsplayed')
plt.ylabel('levelscompleted')
plt.grid()
plt.show()
```

```python
plt.figure()
sns.boxplot(x='platform',y='moneyspent',data=df_plot)
plt.show()
```

```python
sns.pairplot(df_plot[['age','moneyspent','levelscompleted']])
```

```python
plt.figure(figsize=(10,10))
c= df_plot.corr(numeric_only = True)
sns.heatmap(c,cmap="crest",annot=True)
```
