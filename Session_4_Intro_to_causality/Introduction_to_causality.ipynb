** How Do We Know Which Shop Is Better? A/B Testing!**

Entero uses **A/B testing**  a standard experimentation method used by companies like Netflix, Meta, Google, and Bookings to optimize products through data, not opinions.

In business, we often wonder: Did this marketing campaign increase purchases? Or did changing the game interface cause players to play longer?

A/B testing is a scientific way to test these questions  it's a controlled experiment to measure causal impact, not just associations.

<br>

**But, how does A/B Testing work?** 


A/B testing is a controlled experiment where:

-  Users are randomly split into two groups:

| Group | What they see           | What it's called       |
|-------|--------------------------|------------------------|
| A     | The **old** shop design | The **Control** group |
| B     | The **new** shop design | The **Test** group |

-  Each group uses the app/game normally, without knowing they are part of an experiment.

-  We compare the outcomes  did Group B (Test B) buy more? Play longer? Return more often?

-  No switching: A player in Group A stays in A. Same for B. This avoids mixing effects and keeps the test fair and clean.

<br>

** Why A/B Testing = Causality in Action**


If Group B performs better, it's likely because of the new shop !!!!

**Why?** Because since we used random assignment the only difference between the two groups is the shop interface. That's what makes this a causal experiment.

After the running the A/B test experiment, the performance of each of the groups needs to be measured regarding meaningful metrics

** Measuring Performance**

In general one can focus on **key metrics** like:

- **Conversion rate**: % of users who make a purchase
- **Revenue per player** (also called Average Revenue Per User or ARPU)
- **Engagement**: how long players stay in the game or how often they return

and other that might be related to the product change

**Example:**

Imagine that

- In group A (control): 100 out of 1,000 players made a purchase  **10% conversion**
- In group B (test): 130 out of 1,000 players made a purchase  **13% conversion**

This difference is important  and leads us to calculate **uplift**

** Calculating Uplift**

Once we have the results for both groups, we can calculate **uplift**  which tells us how much better (or worse) the new version performed compared to the old one.

**Uplift** is the **percentage increase or decrease** in a key metric (like conversion or revenue) from the control group to the variant group.

**Formula**:

$$\text{Uplift} = \frac{Variant - Control}{Control} \times 100 $$

**Example:**

- Group A (Control): 10% of players made a purchase  
- Group B (Variant): 13% of players made a purchase

$\text{Uplift} = \frac{13\% - 10\%}{10\%} \times 100 = \mathbf{30\%}$

So we say:  
> The new shop increased conversion by **30%** compared to the original version.

** Why does uplift matter?**

Uplift helps us **quantify the impact** of the change. It gives a clear business answer:

- Is the new design **worth keeping**?
- How much more revenue or engagement could we expect if we roll it out to all players?

>   Note that a small uplift can still be **very meaningful** if the player base is large!
